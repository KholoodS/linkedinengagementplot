```python
import pandas as pd
import matplotlib.pyplot as plt

# Specify the path to your Excel file
excel_file = 'Content_2024-03-31_2024-04-06_Kholood JenanSajid.xlsx'

# Load data from Excel into a pandas DataFrame
df = pd.read_excel(excel_file, engine='openpyxl')

# Inspect the structure and columns
print(df.head())

print(df.columns)
```

```python
dates = df['Date']
impressions = df['Impressions']

# Plotting impressions data
plt.figure(figsize=(10, 6))
plt.bar(dates, impressions, color='skyblue')
plt.xlabel('Date')
plt.ylabel('Impressions')
plt.title('LinkedIn Post Impressions')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.show()
```

```python
excel_file = 'job titles.xlsx'

# Load data from Excel into a pandas DataFrame
df = pd.read_excel(excel_file, engine='openpyxl')

print(df.columns)
```

```python
dates = df['Date']
impressions = df['Impressions']
engagements = df['Engagements']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dates, impressions, label='Impressions', marker='o')
plt.plot(dates, engagements, label='Engagements', marker='x')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('LinkedIn Post Metrics Over Time')
plt.xticks(rotation=45)  
plt.legend()
plt.grid(True)
plt.show()
```
